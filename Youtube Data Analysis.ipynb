```python
youtube_links = pd.read_csv("../newYoutubeDoc.csv")
# Convert 'Date' and 'Time' columns to datetime format
youtube_links['Date'] = pd.to_datetime(youtube_links['Date'], format='%d-%b-%y')
#youtube_links['Time'] = pd.to_datetime(youtube_links['Time'])
youtube_links['Time'] = youtube_links['Time'].str.replace(' ', '')
youtube_links['Time'] = pd.to_datetime(youtube_links['Time'])
# Extract hour from the 'Time' column
youtube_links['Hour'] = youtube_links['Time'].dt.hour

# Define time ranges and corresponding colors
time_ranges = {
    (0, 6): 'blue',   # Midnight to 6 AM
    (6, 12): 'green',  # 6 AM to 12 PM
    (12, 18): 'orange',  # 12 PM to 6 PM
    (18, 24): 'red'   # 6 PM to midnight
}

# Plotting date vs time with different colors for different time ranges
plt.figure(figsize=(10, 6))

for (start, end), color in time_ranges.items():
    subset = youtube_links[(youtube_links['Hour'] >= start) & (youtube_links['Hour'] < end)]
    plt.scatter(subset['Date'], subset['Time'], color=color, label=f'{start}-{end} hours', alpha=0.7)

plt.xlabel('Date')
plt.ylabel('Time')
plt.title('Date vs Time for Videos WAtched (Color-coded by Time Range)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
